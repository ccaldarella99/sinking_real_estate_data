# EDA for data from Zillow

All data is Monthly data

```python
import pandas as pd
import numpy as np
import os
```

```python
# get path and file name for downloaded Zillow csv files
zillow_path = '../data/zillow'
zillow_csv_files = os.listdir(zillow_path)
```

```python
zillow_path_and_file = [os.path.join(zillow_path, csv) for csv in zillow_csv_files]
zillow_path_and_file
```

```python
metro_ZillowHomeValueIndex_raw = pd.read_csv(zillow_path_and_file[2])

zhvi_met_raw = metro_ZillowHomeValueIndex_raw.set_index('RegionName').iloc[:,4:]
zhvi_met_raw.head()
```

```python
metro_ZillowHomeValueIndex_smoothed_seasonalAdjusted = pd.read_csv(zillow_path_and_file[3])

zhvi_met_smooth = metro_ZillowHomeValueIndex_smoothed_seasonalAdjusted.set_index('RegionName').iloc[:,4:]
zhvi_met_smooth.head()
```

```python
zillowHomeValueForecasts_raw = pd.read_csv(zillow_path_and_file[5])

zhvf_raw_zip = zillowHomeValueForecasts_raw[zillowHomeValueForecasts_raw['Region'] == 'Zip']
zhvf_raw_zip = zhvf_raw_zip.rename(columns={'RegionName': 'ZipCode'})
zhvf_raw_zip = zhvf_raw_zip.set_index('ZipCode').iloc[:,1:]
# zhvf_raw_zip = zhvf_raw_zip.set_index('RegionName').iloc[:,1:]


zhvf_raw_metro = zillowHomeValueForecasts_raw[zillowHomeValueForecasts_raw['Region'] == 'Msa']
zhvf_raw_metro = zhvf_raw_metro.drop(columns={'CountyName', 'CityName'})
zhvf_raw_metro = zhvf_raw_metro.set_index('RegionName').iloc[:,1:]

print(zhvf_raw_metro.head())
zhvf_raw_zip.head()
```

```python
zillowHomeValueForecasts_smoothed_seasonalAdjusted = pd.read_csv(zillow_path_and_file[6])

zhvf_smooth_zip = zillowHomeValueForecasts_smoothed_seasonalAdjusted[zillowHomeValueForecasts_smoothed_seasonalAdjusted['Region'] == 'Zip']
zhvf_smooth_zip = zhvf_smooth_zip.set_index('RegionName').iloc[:,1:]

zhvf_smooth_metro = zillowHomeValueForecasts_smoothed_seasonalAdjusted[zillowHomeValueForecasts_smoothed_seasonalAdjusted['Region'] == 'Msa']
zhvf_smooth_metro = zhvf_smooth_metro.drop(columns={'CountyName', 'CityName'})
zhvf_smooth_metro = zhvf_smooth_metro.set_index('RegionName').iloc[:,1:]

print(zhvf_smooth_metro.head())
zhvf_smooth_zip.head()
```

```python
zipCode_ZillowHomeValueIndex_smoothed_seasonalAdjusted = pd.read_csv(zillow_path_and_file[4])

zhvi_met_smooth_zip = zipCode_ZillowHomeValueIndex_smoothed_seasonalAdjusted.rename(columns={'RegionName': 'ZipCode'}).set_index('ZipCode').iloc[:,8:]
zhvi_met_smooth_zip.head()
```

```python
zipCode_ZillowHomeValueIndex_smoothed_seasonalAdjusted = pd.read_csv(zillow_path_and_file[7])

zhvi_met_smooth_city = zipCode_ZillowHomeValueIndex_smoothed_seasonalAdjusted.rename(columns={'RegionName': 'ZipCode'})#.set_index('ZipCode')
# zhvi_met_smooth_city = zhvi_met_smooth_city.drop(columns={'RegionID', 'SizeRank', 'RegionType', 'StateName', 'State', 'City', 'Metro', 'CountyName'})
zhvi_met_smooth_city = zhvi_met_smooth_city.drop(columns={'RegionID', 'ZipCode', 'SizeRank', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName'})
zhvi_met_smooth_city = zhvi_met_smooth_city.groupby(['City'])#, 'ZipCode'])

zhvi_met_smooth_city.mean()
```
